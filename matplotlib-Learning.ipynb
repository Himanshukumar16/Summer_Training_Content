```python
from matplotlib import style
style.use('Solarize_Light2')
x = ([2,-2,4],[0,10,5.5])
y = ([23,43,54],[55,66,77])
plt.title("Dummy Data")
plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.plot(x,y)
plt.show()
```

```python
x = ([2,-2,4])
y = ([23,43,54])
a = ([0,10,5.5])
b = [55,66,77]
style.use('ggplot')
plt.title("Dummy")
plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.plot(x,y ,label = 'First')
plt.plot(a,b,label = "Second")
plt.legend()
plt.show()
```

```python
x = ([2,-2,4])
y = ([23,43,54])
a = ([0,10,5.5])
b = [55,66,77]
style.use('ggplot')
plt.title("Dummy")
plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.plot(x,y ,label = 'First',linewidth = 5)
plt.plot(a,b,label = "Second")
plt.legend()
plt.show()
```

```python
x = ([2,-2,4])
y = ([23,43,54])
a = ([0,10,5.5])
b = [55,66,77]
style.use('ggplot')
plt.title("Dummy")
plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.scatter(x,y)
# plt.legend()
plt.show()
```

```python
x = [10,20,30]
y = [50,51,54]
style.use('ggplot')
plt.title("Dummy")
plt.xlabel("Label for X")
plt.ylabel("Label for Y")
plt.plot(x)
plt.hist(x,linewidth = 5)
# plt.legend()
plt.show()
```
